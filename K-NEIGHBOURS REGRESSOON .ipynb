```python
import pandas as pd
import seaborn as sna
```

```python
data=pd.read_csv("salary.csv")
```

```python
data.head(3)
```

```python
data.isnull().sum()
```

```python
data["Age"]=data["Age"].fillna(data["Age"].mean())
```

```python
data["Years of Experience"]=data["Years of Experience"].fillna(method="ffill")
```

```python
data["Salary"]=data["Salary"].fillna(data["Salary"].mean())
```

```python
data.isnull().sum()
```

```python
x=data.iloc[:,:-1]
y=data["Salary"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor
kr=KNeighborsRegressor(n_neighbors=4)
```

```python
kr.fit(x_train,y_train)
```

```python
kr.score(x_train,y_train)*100
```

```python
kr.score(x_test,y_test)*100
```

```python
for i in range(1,30):
    kr1=KNeighborsRegressor(i)
    kr1.fit(x_train,y_train)
    print(i,kr1.score(x_train,y_train)*100,kr1.score(x_test,y_test)*100)
```
